```python
# feature engineering 
# chúng ta tạo ra các feature từ các feature sẵn có --> sinh ra thêm các feature 

#weekday --> sẽ phản ánh giá của ngày tiếp 
# feature_01 ---> weekday của cột date
```

```python
df_train['feat_01'] = df_train['Date'].dt.weekday
```

```python
df_train
#Monday=0, Sunday=6
#feat_01 là numeric nhưng mang tính chất là categorical
```

```python
# đổi định dạng của cột dữ liệu sang type khác (int -> str)
df_train['feat_01'] = df_train['feat_01'].astype('str')
```

```python
df_train.info()
```

```python
#feature representation
# cần thêm feature nào
```

```python
# get the average "Close" price of the most recent 5 days
Prev_5_avg = []
for index, row in df_train.iterrows():
    if index < 5: 
        Prev_5_avg.append(0)
    else:
        i_5 = index - 5
        tmp = df_train.iloc[i_5:index]['Close'].mean()
        Prev_5_avg.append(tmp)
```

```python
# 21/2/2021 ----> 21/2/2021
# giá trung bình của 5 ngày trước đó .... không ai trả lời được, feature engineering --> thử sai
```

```python
# df['Prev_5_avg'] = Prev_5_avg
df_train['feat_02'] = Prev_5_avg
```

```python
df_train
```

```python
#feat_03 ---- Prev_30_avg
Prev_30_avg = []
for index, row in df_train.iterrows():
    if index < 30: 
        Prev_30_avg.append(0)
    else:
        i_30 = index - 30
        tmp = df_train.iloc[i_30:index]['Close'].mean()
        Prev_30_avg.append(tmp)

#shift ---> lead, log trong sql --> window function ---> run nhanh hơn hẳn cách trên
```

```python
df_train['feat_03'] = Prev_30_avg
```

```python
df_train
# feature theo mean
# standard deviation std
```

```python
#get the std of the "close" price of the most recent 5 and 30 days:
Prev_5_std = []
for index, row in df_train.iterrows():
    if index < 5: 
        Prev_5_std.append(0)
    else:
        i_5 = index - 5
        tmp = df_train.iloc[i_5:index]['Close'].std()
        Prev_5_std.append(tmp)
df_train['feat_04'] = Prev_5_std

Prev_30_std = []
for index, row in df_train.iterrows():
    if index < 30: 
        Prev_30_std.append(0)
    else:
        i_30 = index - 30
        tmp = df_train.iloc[i_30:index]['Close'].std()
        Prev_30_std.append(tmp)
df_train['feat_05'] = Prev_30_std
```

```python
df_train
```

```python
# khai thác dữ liệu về giá (ở phía trên)
# volume ---> khối lượng cảm nhận sẽ ảnh hưởng tới giá
#khối lượng giao dịch trung bình của 5 ngày gần nhất 
#khối lượng giao dịch trung bình của 30 ngày gần nhất 

Prev_volume_5_std = []
Prev_volume_5_avg = []
for index, row in df_train.iterrows():
    if index < 5: 
        Prev_volume_5_std.append(0)
        Prev_volume_5_avg.append(0)
    else:
        i_5 = index - 5
        
        #get std
        tmp = df_train.iloc[i_5:index]['Volume'].std()
        Prev_volume_5_std.append(tmp)
        
        #get mean
        tmp = df_train.iloc[i_5:index]['Volume'].mean()
        Prev_volume_5_avg.append(tmp)
        
df_train['feat_06'] = Prev_volume_5_std
df_train['feat_07'] = Prev_volume_5_avg

#get the average and std of volume of the last 30 days
Prev_volume_30_std = []
Prev_volume_30_avg = []
for index, row in df_train.iterrows():
    if index < 30: 
        Prev_volume_30_std.append(0)
        Prev_volume_30_avg.append(0)
    else:
        i_30 = index - 30
        
        #get std
        tmp = df_train.iloc[i_30:index]['Volume'].std()
        Prev_volume_30_std.append(tmp)
        
        #get mean
        tmp = df_train.iloc[i_30:index]['Volume'].mean()
        Prev_volume_30_avg.append(tmp)
        
df_train['feat_08'] = Prev_volume_30_std
df_train['feat_09'] = Prev_volume_30_avg
```

```python
# tạm hài lòng với đống feature engineering 
df_train_01 = df_train[['feat_01', 'feat_02', 'feat_03', 'feat_04', 'feat_05', 'feat_06', 'feat_07', 'feat_08', 'feat_09', 'Open']]
```

```python
#x_train, y_train 
#x_train: những feature dùng để huấn luyện mô hình 
# y_train: biến mục tiêu, target variable, label

X_train = df_train_01[['feat_01', 'feat_02', 'feat_03', 'feat_04', 'feat_05', 'feat_06', 'feat_07', 'feat_08', 'feat_09']]
Y_train = df_train_01[['Open']]
```
